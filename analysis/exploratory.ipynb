# COVID-19 Tracker - Exploratory Data Analysis (EDA)

This notebook is used for exploratory analysis of the Johns Hopkins CSSE and Our World in Data datasets. It is not run as part of the production Flask web server.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Load Raw JHU Confirmed Cases

```python
CONFIRMED_URL = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
df_raw = pd.read_csv(CONFIRMED_URL)
print(f"Shape of raw confirmed cases: {df_raw.shape}")
df_raw.head()
```

## 2. Melt Wide format to Long format

```python
id_vars = ["Province/State", "Country/Region", "Lat", "Long"]
df_long = df_raw.melt(id_vars=id_vars, var_name="Date", value_name="Cases")
df_long["Date"] = pd.to_datetime(df_long["Date"])
df_long = df_long.groupby(["Country/Region", "Date"])["Cases"].sum().reset_index()
df_long.head()
```

## 3. Visualize Cases for specific countries

```python
target_countries = ["US", "India", "Brazil", "United Kingdom"]
df_filtered = df_long[df_long["Country/Region"].isin(target_countries)]

sns.lineplot(data=df_filtered, x="Date", y="Cases", hue="Country/Region")
plt.title("Cumulative Confirmed Cases comparison")
plt.ylabel("Cumulative Cases")
plt.yscale('log')  # Log scale for exponential comparison
plt.show()
```
